```python
import kagglehub
import pandas as pd
import os
import glob

# 1. Download/Locate the dataset
path = kagglehub.dataset_download("gowrishankarp/newspaper-text-summarization-cnn-dailymail")
print(f"Base Path: {path}")

# 2. Automatically find 'train.csv' no matter where it's hidden
# We use '**/*.csv' to search all subfolders
csv_files = glob.glob(os.path.join(path, "**", "train.csv"), recursive=True)

if not csv_files:
    print("❌ Still can't find train.csv. Let's see what's actually there:")
    print(os.listdir(path))
else:
    target_file = csv_files[0]
    print(f"✅ Found file at: {target_file}")
    
    # 3. Load a small chunk to verify
    train_df = pd.read_csv(target_file, nrows=100)
    print("\nSuccess! Data Preview:")
    print(train_df[['article', 'highlights']].head(2))
```

```python
def clean_text(text):
    # Remove the "By . [Source] . PUBLISHED:" prefix common in this dataset
    if "PUBLISHED:" in text:
        text = text.split("PUBLISHED:")[1]
    
    # Basic cleanup: remove extra newlines and spaces
    text = text.replace("\n", " ").strip()
    return text

# Apply to our small sample
train_df['article_clean'] = train_df['article'].apply(clean_text)

print("Original:", train_df['article'].iloc[0][:100])
print("Cleaned:", train_df['article_clean'].iloc[0][:100])
```

```python
# Count words in each article
train_df['word_count'] = train_df['article_clean'].apply(lambda x: len(x.split()))

print(f"Average word count: {train_df['word_count'].mean()}")
print(f"Max word count: {train_df['word_count'].max()}")
```
